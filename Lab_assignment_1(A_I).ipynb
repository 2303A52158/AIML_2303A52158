**Part 2 – Visualization of Data using matplotlib, pyplots Packages[CO2]**


**Genarate Scatterplot**

**Let's Suppose we colleted age(in years),height(in cm),and weight(in kiolgarms)information from few patients visiting a clinic.boad text**

```python
people=['Ann','Brandon','Chen','David','Emily','Farook',
        'Gagan','Hamish','Imran','Julio','Katherine','Lily']
age=[21,12,32,45,37,18,28,52,5,40,48,15]
weight=[55,35,77,68,70,60,72,69,18,65,82,48]
height=[10,135,170,165,173,168,175,159,105,171,155,158]
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relation between Weight and height of patients" )
plt.ylabel("height in meters")
plt.xlabel("Weight of patiens in kg")
plt.show()
```

**Genarate Barplot**

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')
#Main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,
        color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
#Main plot function 'hist'
plt.hist(weight,color='blue',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
